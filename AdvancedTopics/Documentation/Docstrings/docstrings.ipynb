### Docstrings

```python
def hello(arg):
    """This is the docstring of the function"""
    print("Hello", arg, "!")
```

```python
hello("Toto")
```

```python
help(hello)
```

```python
class ClassExample:
    """This is the docstring of the class"""
    def __init__(self):
        """This is the docstring of the initializator of the class"""
        pass
    
    def method(self):
        """This is the docstring of the method of the class"""
        pass
```

```python
c = ClassExample()
```

```python
help(c)
```
