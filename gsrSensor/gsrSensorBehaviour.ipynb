```python
import pandas as pd 
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
mainpath = "/Users/avega/Documents/Developer/gsrSensor/"
FuruiAfterNoSuit = "resistance.csv"
# import os has a function to join the strings 
fullpath = os.path.join(mainpath, FuruiAfterNoSuit)
# Use the method read_csv() to load the data seat in the variable 
resistanceVoltageData = pd.read_csv(fullpath)
resistanceVoltageData.head()
```

```python
# Change the columns to a list indicating the name of the column
noResistance_list = resistanceVoltageData["No resistence"].tolist()
tenKVoltage_list = resistanceVoltageData["10K voltage"].tolist()
fiftyOneKVoltage_list = resistanceVoltageData["51K voltage"].tolist()
oneHundredKVoltage_list = resistanceVoltageData["100K voltage"].tolist()
twoHundredKVoltage_list = resistanceVoltageData["200K voltage"].tolist()
fiveHundredTenKVoltage_list = resistanceVoltageData["510K voltage"].tolist()
sevenHundredfiftyKVoltage_list = resistanceVoltageData["750K voltage"].tolist()
oneMVoltage_list = resistanceVoltageData["1M voltage"].tolist()
twoPointTwoMVoltage_list = resistanceVoltageData["2.2M voltage"].tolist()

# Convert to array
noResistance = np.array(noResistance_list)
tenKVoltage = np.array(tenKVoltage_list)
fiftyOneKVoltage = np.array(fiftyOneKVoltage_list)
oneHundredKVoltage = np.array(oneHundredKVoltage_list)
twoHundredKVoltage = np.array(twoHundredKVoltage_list)
fiveHundredTenKVoltage = np.array(fiveHundredTenKVoltage_list)
sevenHundredfiftyKVoltage = np.array(sevenHundredfiftyKVoltage_list)
oneMVoltage = np.array(oneMVoltage_list)
twoPointTwoMVoltage = np.array(twoPointTwoMVoltage_list)

# Calculate the average
noResistance_mean = np.mean(noResistance)
tenKVoltage_mean = np.mean(tenKVoltage)
fiftyOneKVoltage_mean = np.mean(fiftyOneKVoltage)
oneHundredKVoltage_mean = np.mean(oneHundredKVoltage)
twoHundredKVoltage_mean = np.mean(twoHundredKVoltage)
fiveHundredTenKVoltage_mean = np.mean(fiveHundredTenKVoltage)
sevenHundredfiftyKVoltage_mean = np.mean(sevenHundredfiftyKVoltage)
oneMVoltage_mean = np.mean(oneMVoltage)
twoPointTwoMVoltage_mean = np.mean(twoPointTwoMVoltage)

# Calculate the std
noResistance_std = np.std(noResistance)
tenKVoltage_std = np.std(tenKVoltage)
fiftyOneKVoltage_std = np.std(fiftyOneKVoltage)
oneHundredKVoltage_std = np.std(oneHundredKVoltage)
twoHundredKVoltage_std = np.std(twoHundredKVoltage)
fiveHundredTenKVoltage_std = np.std(fiveHundredTenKVoltage)
sevenHundredfiftyKVoltage_std = np.std(sevenHundredfiftyKVoltage)
oneMVoltage_std = np.std(oneMVoltage)
twoPointTwoMVoltage_std = np.std(twoPointTwoMVoltage)
```

```python
plt.plot(noResistance, label='No resistence')
plt.plot(tenKVoltage, label='10K voltage')
plt.plot(fiftyOneKVoltage, label='51K voltage')
plt.plot(oneHundredKVoltage, label='100K voltage')
plt.plot(twoHundredKVoltage, label='200K voltage')
plt.plot(fiveHundredTenKVoltage, label='510K voltage')
plt.plot(sevenHundredfiftyKVoltage, label='750K voltage')
plt.plot(oneMVoltage, label='1M voltage')
plt.plot(twoPointTwoMVoltage, label='2.2M voltage')
plt.xlabel('Number of samples')
plt.ylabel('GSR sensor voltage readings')
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
plt.show()
```

```python
# Create lists for the plot
Conditions = ['No resistance', '10K voltage', '51K voltage', '100K voltage', '200K voltage', 
              '510K voltage', '750K voltage', '1M voltage', '2.2M voltage']
x_pos = np.arange(len(Conditions))
BPM = [noResistance_mean, tenKVoltage_mean, fiftyOneKVoltage_mean, oneHundredKVoltage_mean, twoHundredKVoltage_mean,
      fiveHundredTenKVoltage_mean, sevenHundredfiftyKVoltage_mean, oneMVoltage_mean, twoPointTwoMVoltage_mean]
error = [noResistance_std, tenKVoltage_std, fiftyOneKVoltage_std, oneHundredKVoltage_std, twoHundredKVoltage_std,
      fiveHundredTenKVoltage_std, sevenHundredfiftyKVoltage_std, oneMVoltage_std, twoPointTwoMVoltage_std]

# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, BPM, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('GSR sensor voltage readings average')
ax.set_xticks(x_pos)
ax.set_xticklabels(Conditions)
ax.set_title('Cases of study')
ax.yaxis.grid(True)

plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')

# Save the figure and show
plt.tight_layout()
# plt.savefig('bar_plot_with_error_bars.png')
plt.show()
```
